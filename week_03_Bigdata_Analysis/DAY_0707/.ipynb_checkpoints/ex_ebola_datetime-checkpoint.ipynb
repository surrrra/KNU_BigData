```python
# 모듈 로딩
import pandas as pd

# 파일 관련 변수들
DIR='../Data/'
FILE=DIR+'country_timeseries.csv'
```

```python
# (1) CSV => DataFrame 로딩
ebolaDF=pd.read_csv(FILE)
ebolaDF
```

```python
# (2) 정보 확인
ebolaDF.info()
```

```python
# (3) 전처리
# (3-1) NaN 데이터 처리 =>
ebolaDF.fillna(0, inplace=True)
ebolaDF.isnull().sum()
```

```python
ebolaDF.head()
```

```python
ebolaDF.info()
```

```python
# (3-2) Data Type 정리
ebolaDF['Date']=pd.to_datetime(ebolaDF['Date'])
```

```python
ebolaDF.info()
```

###### 에볼라 바이러스가 발병한지 며칠이 지났는지 컬럼 추가

```python
# 최초 발병일로부터 Date까지 얼마나 지났는지 계산
ebola_min=ebolaDF['Date'].min()
ebolaDF['diff']=ebolaDF['Date']-ebola_min
```

```python
ebolaDF
```

```python
# 발병일로부터 오늘까지의 날짜 계산
from datetime import datetime
now=pd.to_datetime(datetime.now().strftime('%Y-%m-%d'))
# strtfime('%Y-%m-%d')를 안쓰면 시간까지 출력돼서 그냥 년-월-일만 뽑았음
```

```python
datetime.now()-ebolaDF['Date']   # strftime('%Y-%m-%d')를 안 썼을 때
```

```python
now
```

```python
ebolaDF['now_diff']=now-ebolaDF['Date']
```

```python
ebolaDF
```

```python
ebolaDF.info()
```

```python
ebolaDF['now_diff'].dt.days  # 얘를 쓰면 int 형태로 나옴
```

```python
ebolaDF['now_diff'].dtype
```

```python
now_diff2=now-ebolaDF['Date']
now_diff2[0].days    # 속성 찾아보는걸 잊지말자
```
