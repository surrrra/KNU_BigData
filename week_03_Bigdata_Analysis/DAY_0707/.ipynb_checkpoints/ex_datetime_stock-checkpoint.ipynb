# Datetime => Index로 설정 실습

- set_index(컬럼) : 인덱스 설정

```python
# 모듈 로딩
import pandas as pd

# 파일 관련 변수들
DIR='../Data/'
FILE=DIR+'stock-data.csv'
```

```python
# (1) CSV FILE => DataFrame 로딩
stockDF=pd.read_csv(FILE)
```

```python
stockDF.info()
```

```python
# (2) 데이터 확인
stockDF.head()
```

```python
stockDF.isnull().sum()
```

```python
# (3) 전처리 => NaN, Duplicate, DateType
# Date 컬럼 Object => datetime64

# 방법 (1) series.astype()
# 데이터 타입을 datetime으로 바꾸는 또다른 방법
stockDF['Date']=stockDF['Date'].astype('datetime64')
```

```python
# 방법 (2) pandas.to_datetime()
stockDF['Date']=pd.to_datetime(stockDF['Date'])
stockDF.info()
```

```python
stockDF.head()
```

```python
# 특정 데이터 추출
# 7월 데이터만 뽑아내기
stockDF[stockDF['Date'].dt.month==7]
```

```python
# 6월 20일부터 25일까지의 데이터
stockDF[(stockDF['Date'].dt.month==6) & (stockDF['Date'].dt.day>=20) & (stockDF['Date'].dt.day<=25)]
```

```python
# Index 변경 => Date 컬럼을 인덱스로 설정 => set_index(컬럼명)
stockDF.set_index('Date', inplace=True)
```

```python
stockDF
```

```python
stockDF.info()
```

```python
stockDF.index
```

```python
stockDF[(stockDF.index.month==6)&(stockDF.index.day>=20)&(stockDF.index.day<=25)]
```

```python
stock_month=stockDF.index.month
stock_day=stockDF.index.day
#stockDF.loc[stockDF.index.month==6]
stockDF.loc[(stock_month==6)&(stock_day>=20)&(stock_day<=25)]  # 근데 loc 안써도 나온다
# 파이참에서 돌리면 경고가 뜬다. 그니까 그냥 행 추출할 때는 loc 넣는게 마음 편함
```
