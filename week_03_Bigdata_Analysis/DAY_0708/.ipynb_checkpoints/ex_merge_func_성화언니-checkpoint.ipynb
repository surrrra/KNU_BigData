## DF & SR 통합 메서드 실습 --------------------------

- pandas.merge() / DataFrame.join() 제공

```python
# 모듈 로딩 -----------------------------
import pandas as pd

# 파일 관련 변수들 ----------------------
DIR='../Data/'
FILE1=DIR+'stock price.xlsx'
FILE2=DIR+'stock valuation.xlsx'
```

```python
# 데이터 로딩 -----------------------------
df1=pd.read_excel(FILE1)
df2=pd.read_excel(FILE2)
```

```python
df1.head()
```

```python
df2.head()
```

# DF 합치기 => merge ----------------------

### [기본] DF 사이의 공통된 컬럼명을 기준으로 교집합(join) 합치기

```python
allDF=pd.merge(df1,df2)
allDF
# id 컬럼을 기준으로 겹치는 것만 나타남
```

##### [수정] 2개 DF의 모든 공통 컬럼 기준으로 모든 데이터 합치기(outer)

```python
allDF2=pd.merge(df1,df2, how='outer')
allDF2
```

[수정] 2개 DF에서 왼쪽(Left) 기준으로 데이터 합치기(Left Outer Join)

```python
allDF3=pd.merge(df1,df2, how='left')
allDF3
# 왼쪽을 기준으로 만들어서, 왼쪽 값은 다 있음
# id를 기준으로 하여 id는 1개 열만 있음
```

```python
allDF3=pd.merge(df1,df2, how='left', left_on='stock_name', right_on='name')
allDF3
# 네임이 같은걸로 보기 때문에 각각 id 출력되어 2가지 있음(id_x, id_y)
```

[수정] 2개 DF에서 오른쪽(Right) 기준으로 데이터 합치기(Right Outer Join)

```python
allDF4=pd.merge(df1,df2, how='right', left_on='stock_name', right_on='name')
allDF4
```

### [DataFrame 메서드] join ---------------------------------------------

```python
df1.join(df2, lsuffix='_left', rsuffix='_right')
```
