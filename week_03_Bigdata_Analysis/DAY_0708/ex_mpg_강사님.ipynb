## (3-3) 데이터 정규화

- 수치데이터 컬럼마다 값의 범위가 다름
- 값의 범위 0 ~ 1 or -1 ~ 1로 값을 설정

### (3-3-1) 방법1 : 컬럼의 최대값의 절댓값으로 전체 데이터를 나누기

```python
mpg['weight']
```

```python
mpg['weight_nor1']=mpg['weight']/abs(mpg['weight'].max())
```

```python
mpg['weight_nor1'].min(), mpg['weight_nor1'].max()
```

### (3-3-2) 방법 : 컬럼의 데이터 - 최소값/최대값-최소값

- 컬럼의 최대값 => max(), 최소값 => min()

```python
# 분자
numerator = mpg['weight'] - mpg['weight'].min()
# 분모
denominator=mpg['weight'].max()-mpg['weight'].min()

mpg['weight_nor_2']=numerator/denominator
mpg['weight_nor_2']
```

```python
mpg['weight_nor_2'].min(), mpg['weight_nor_2'].max()
```

### (3-3-3) 방법 : 컬럼의 데이터 - 평균/표준편차  => Z-score

```python
weight_mean=mpg['weight'].mean()
weight_std=mpg['weight'].std()
mpg['weight_nor_3']=(mpg['weight']-weight_mean)/weight_std
```

```python
mpg['weight_nor_3'].min(), mpg['weight_nor_3'].max()
```

```python
mpg['weight_nor_3']
```

```python
mpg[['weight_nor1', 'weight_nor_2', 'weight_nor_3']]
```

```python
def new_nor():
    for col in mpg.columns:
        if mpg[col].dtype in ['float', 'int64']:
            numerator=mpg[col]-mpg[col].min()
            denominator=mpg[col].max()-mpg[col].min()
            mpg[col+'_nor']=numerator/denominator

new_nor()
```

```python
# 저대로 했더니 정규화 시킨 컬럼도 정규화가 돼서 컬럼을 drop했음
mpg.drop(columns=['weight_nor1_nor', 'weight_nor_2_nor', 'weight_nor_3_nor'], inplace=True)
```
