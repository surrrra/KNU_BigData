# k-NN 알고리즘 score 확인

### k-NN 알고리즘의 neighbor 개수에 따른 score 값 변화 확인
- n_neighbors의 값을 5부터 49까지 바꾸어 가면서 score() 함수의 결과 및 
- n_neighbors의 값을 plot 차트로 표현하시오.

```python
# 도미 데이터
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 
                30.7, 31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 
                33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 
                36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]

bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 
                450.0, 500.0, 475.0, 500.0, 500.0, 340.0, 600.0, 600.0, 
                700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 
                925.0, 975.0, 950.0]

# 빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 
                12.2, 12.4, 13.0, 14.3, 15.0]

smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 
                13.4, 12.2, 19.7, 19.9]
```

```python
length=bream_length+smelt_length
weight=bream_weight+smelt_weight

fish_data=[[l, w] for l, w in zip(length, weight)]
print(fish_data)

# 혹은
# fish_data=np.column_stack((fish_length, fish_weight))도 가능
```

```python
fish_target=[1]*35+[0]*14
print(fish_target)

# 혹은
# fish_target=np.concatenate((np.ones(35), np.zeros(14)))도 가능
```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np

n_neighbors_list=[]

for i in range(5, 50):
    
    kn=KNeighborsClassifier(n_neighbors=i)
    kn.fit(fish_data,fish_target)

    n_neighbors_list.append(kn.score(fish_data, fish_target))

plt.figure(figsize=(20, 8))
plt.xlabel('neighbors')
plt.ylabel('score')
plt.plot(n_neighbors_list, marker='s')
plt.xticks(range(45), range(5, 50))
plt.show()
```

## plot()에 사용한 n_neighbors와 score 값을 이용하여  
## n_neighbors의 개수에 대한 score 값을 DataFrame으로 변환하여 출력

```python
import pandas as pd
neighbors=[i for i in range(5, 50)]

my_data=np.column_stack((neighbors, n_neighbors_list))
df=pd.DataFrame(my_data, columns=['neighbors','score'])
df
```
