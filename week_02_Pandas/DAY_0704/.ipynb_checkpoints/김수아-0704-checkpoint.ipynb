```python
import pandas as pd

# 파일 관련 변수 선언
DIR_PATH='../Data/'
FILE_NAME=DIR_PATH+'weather.csv'
```

```python
# (1) CSV FILE => DataFrame으로 로딩
weather=pd.read_csv(FILE_NAME)
```

```python
# (2) 데이터 정보 확인
weather.head()
```

```python
# 과제 0704-(1) 2010년 1월~12월까지 최고온도/최저온도 추출
i=0
j=1
while i <= 21:
    if i!=16:  # 9월이 없음
        weather_max=weather.iloc[[i, i+1]].dropna(axis=1, how='all').iloc[:,4:].iloc[0].max()
        weather_min=weather.iloc[[i, i+1]].dropna(axis=1, how='all').iloc[:,4:].iloc[1].min()
        print(f'{j}월의 최고온도 => {weather_max}')
        print(f'{j}월의 최저온도 => {weather_min}')
        i+=2
        j+=1
    else: 
        j+=1
        weather_max=weather.iloc[[i, i+1]].dropna(axis=1, how='all').iloc[:,4:].iloc[0].max()
        weather_min=weather.iloc[[i, i+1]].dropna(axis=1, how='all').iloc[:,4:].iloc[1].min()
        print(f'{j}월의 최고온도 => {weather_max}')
        print(f'{j}월의 최저온도 => {weather_min}')
        i+=2
        j+=1
```
