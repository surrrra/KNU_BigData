# 2번 문제

```python
import pandas as pd


iris=pd.read_csv('../Data/iris.csv')
iris
```

## 2-1번

```python
iris[['SepalLength', 'PetalLength']]
```

```python
# 나중에 데이터가 엄청나게 커진다면 ,, 자동화

col_name=[]
for col in iris.columns:
    #print(col[-6:])
    if col[-6:]=='Length':
        col_name.append(col)

print(pd.DataFrame(iris[col_name]))
```

## 2-2번

```python
iris[iris['SepalLength']>=5.0][['SepalLength', 'Name']]
```
