### ----------------------------------- 알루미늄 가격 동향 -----------------------------------

#### 출처 : https://www.kores.net/komis/price/mineralprice/basemetals/pricetrend/baseMetals.do

```python
import pandas as pd
import matplotlib.pyplot as plt
import platform

aluminum=pd.read_excel('알루미늄.xls')
aluminum.head()
```

```python
# 필요한 컬럼만 뽑아오기
aluminum=aluminum[['기준일', '기준가격']]
aluminum.head()
```

```python
# 요약정보 확인
aluminum.info()
```

```python
# 컬럼 자료형 변환
aluminum['기준일']=pd.to_datetime(aluminum['기준일'])
```

```python
aluminum.info()
```

```python
aluminum['month']=aluminum['기준일'].dt.month
```

```python
aluminum['year']=aluminum['기준일'].dt.year
```

```python
aluminum
```

```python
al_group=pd.DataFrame(aluminum.groupby(['year', 'month']).mean())
al_group
```

### --------------------------------------------- 구리 ---------------------------------------------

#### 출처 : https://www.index.go.kr/potal/main/EachDtlPageDetail.do?idx_cd=1378

```python
guri=pd.read_excel('구리.xls', thousands=',', skiprows=2)
guri
```

### ----------------------------------- 소비자물가지수 동향 -----------------------------------

#### 출처 : https://kosis.kr/statHtml/statHtml.do?orgId=101&tblId=DT_1J20003&checkFlag=N

```python
gisu=pd.read_csv('소비자물가지수.csv', encoding='CP949', index_col=0)
gisu
```

```python
gisu.rename(columns={'2020.1':'2020.10', '2021.1':'2021.10'}, inplace=True)
```

```python
gisu=gisu.T
gisu
```

```python
# 날짜를 컬럼으로 빼기 위해서
gisu.reset_index(inplace=True)
gisu
```

```python
gisu.info()
```

```python
# 컬럼 이름 변경
gisu.columns=['기준일', '소비자물가지수']
```

```python
gisu['year']=gisu['기준일'].str.split('.').str[0].astype('int')
gisu['month']=gisu['기준일'].str.split('.').str[1].astype('int')
```

```python
gisu
```

```python
gisu.info()
```

```python
gisu_group=gisu.groupby(['year', 'month']).mean()
gisu_group
```

## 2. 스케일링

```python
# 정규화 하는 함수
def get_scale(df_name):
    return (df_name['기준가격']/df_name['기준가격'].max()).tolist()
```

### 알루미늄

```python
al_list=get_scale(al_group)
al_list
```

### 게르마늄

```python
ger_list=get_scale(ger_group)
ger_list
```

```python
ger_scale_mean=sum(ger_list)/len(ger_list)
ger_scale_mean
```

```python
# ger 파일에 2020년 9월이 누락되어있음 -> 평균값으로 대치
ger_list.insert(8, ger_scale_mean)
```

```python
ger_list
```

### 금

```python
gold_list=get_scale(gold_group)
gold_list
```

### 구리

```python
guri_list=get_scale(guri_group)
guri_list
```

### 소비자물가지수

```python
gisu_list=(gisu_group['소비자물가지수']/gisu_group['소비자물가지수'].max()).tolist()
gisu_list
```

## 그래프

```python
# x축 라벨을 위한 코딩
x_ticks=[]
for i in range(2020, 2022):
    for j in range(1, 13):
        x_ticks.append(str(i)+'-'+str(j))

for i in range(1, 7):
    x_ticks.append('2022-'+str(i))
                       
x_ticks
```

```python
# 다른 운영체제에서도 적용할 수 있도록 한글 폰트 깨짐 방지
if platform.system()=='Windows':
    plt.rc('font', family='Malgun Gothic', size=8)
else:
    plt.rc('font', family='AppleGothic', size=8)
```

```python
plt.figure(figsize=(20, 8))
plt.xticks(rotation=80)
ax1=plt.subplot(111)
ax2=ax1.twinx()
a=ax1.plot(x_ticks, al_list, label='알루미늄', marker='o', ms=4)
a=ax1.plot(x_ticks, ger_list, label='게르마늄', marker='o', ms=4)
a=ax1.plot(x_ticks, gold_list, label='금', marker='o', ms=4)
a=ax1.plot(x_ticks, guri_list, label='구리', marker='o', ms=4)
b=ax2.plot(x_ticks, gisu_list, label='소비자물가지수', color='purple',marker='o', ms=4)

ax1.set_ylabel('반도체 원료 가격')
ax2.set_ylabel('소비자물가지수')

ax1.legend()
ax2.legend(loc='lower right')

plt.show()
```

소비자물가지수 : 소비자가 구입하는 상품과 서비스의 가격 변동을 측정하기 위한 지표

알루미늄, 게르마늄, 구리 : 소비자물가지수에 영향을 받는 것으로 보임.

### 상관계수 구하기

```python
def get_corr(list_name):
    corr_list=pd.DataFrame(list_name, gisu_list)
    corr_list.reset_index(inplace=True)
    return corr_list
```

#### 소비자물가지수와 알루미늄의 상관관계

```python
corr_al=get_corr(al_list)
corr_al.columns=['소비자물가지수', '알루미늄']
corr_al.corr()
```

#### 소비자물가지수와 게르마늄의 상관관계

```python
corr_ger=get_corr(ger_list)
corr_ger.columns=['소비자물가지수', '게르마늄']
corr_ger.corr()
```

#### 소비자물가지수와 금의 상관관계

```python
corr_gold=get_corr(gold_list)
corr_gold.columns=['소비자물가지수', '금']
corr_gold.corr()
```

#### 소비자물가지수와 구리의 상관관계

```python
corr_guri=get_corr(guri_list)
corr_guri.columns=['소비자물가지수', '구리']
corr_guri.corr()
```

금은 인플레이션과 큰 관계는 없지만
