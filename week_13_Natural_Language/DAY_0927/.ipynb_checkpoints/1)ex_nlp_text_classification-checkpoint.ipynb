```python
words=[]

for sent in newsData20:
    words.append(word_tokenize(sent))
```

```python
len(words)
```

```python
len(words[0])
```

```python
en_stopwords=nltk.corpus.stopwords.words('english')
```

```python
# 불용어 제거
  
for i in range(len(words)):
    for w in words[i]:
        if w not in en_stopwords:
            words[i].remove(w)
```
