## HTML 코드 확인
- 매장 정보 확인
    - '< t b o d y >' 태그 내부
        - '< t d >' 태그 첫 번째 : 매장이 있는 지역
        - '< t d >' 태그 두 번째 : 매장 이름
        - '< t d >' 태그 네 번째 : 매장 주소
        - '< t d >' 태그 다섯 번째 : 매장 전화번호
- 나머지 매장 정보 확인
    - https://www.hollys.co.kr/store/korea/korStore2.do?pageNo=1&sido=&gugun=&store=

### 1. 매장 찾기에서 1~53페이지까지 모든 매장의 정보를 스크레이핑
- 지역, 매장명, 매장 주소, 전화번호
- 수집된 정보는 Pandas DataFrame으로 변경한 다음 csv 파일로 저장
- csv 파일 : hollys.branches.csv
- 소스 파일 : hw02.coffeeshop_to_csv.ipynb

### 2. 저장된 csv 파일을 읽어서 사용자가 입력한 지역에 있는 매장 정보 출력
- 소스 파일 : hw02_find_coffeeshop.ipynb

```python
tds=[]
result=[]

region=[]     # 지역을 담는 리스트
shop_name=[]  # 매장 이름을 담는 리스트
shop_addr=[]  # 매장 주소를 담는 리스트
call=[]       # 전화번호를 담는 리스트
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

tds=[]     # <td> 태그 담는 리스트
result=[]  # <td> 태그의 text 담는 리스트

region=[]     # 지역을 담는 리스트
shop_name=[]  # 매장 이름을 담는 리스트
shop_addr=[]  # 매장 주소를 담는 리스트
call=[]       # 전화번호를 담는 리스트
```

```python
def get_info(soup):
    '''
    지역, 매장이름, 매장주소, 전화번호를 리스트에 담는 함수
    '''
    tbodys=soup.find('tbody')
    tbody=tbodys.find_all('td')
    
    for t in tbody:
        tds.append(t.text)
```

```python
def seperate_result(result):
    '''
    모든 정보가 담겨있는 리스트를
    지역
    매장이름
    매장주소
    전화번호
    총 4개의 리스트에 각각 나누어 담는 함수
    '''
    
    for i in range(len(result)):
        if i%4==0:
            region.append(result[i])
        elif i%4==1:
            shop_name.append(result[i])
        elif i%4==2:
            shop_addr.append(result[i])
        else:
            call.append(result[i])
```

```python
def get_df():
    '''
    리스트를 하나의 데이터프레임으로 만드는 함수
    '''
    region_df=pd.DataFrame(region, columns=['지역'])
    shop_name_df=pd.DataFrame(shop_name, columns=['매장이름'])
    shop_addr_df=pd.DataFrame(shop_addr, columns=['매장주소'])
    call_df=pd.DataFrame(call, columns=['전화번호'])
    info=pd.concat([region_df, shop_name_df, shop_addr_df, call_df], axis=1)
    return info
```

```python
for i in range(1, 3):
    html=urlopen('https://www.hollys.co.kr/store/korea/korStore2.do?pageNo='+str(i)+'&sido=&gugun=&store=')
    soup=BeautifulSoup(html, 'html.parser')

    get_info(soup)
    
#     print(i,'번째 for문 돌아가는중 ...', result)
#     print()
    
for i in range(1, len(tds)+1):
    if (i%6  in [0, 1, 2, 4]):
        result.append(tds[i-1])

seperate_result(result)

info=get_df()
pd.DataFrame(info)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

tds=[]     # <td> 태그 담는 리스트
result=[]  # <td> 태그의 text 담는 리스트

region=[]     # 지역을 담는 리스트
shop_name=[]  # 매장 이름을 담는 리스트
shop_addr=[]  # 매장 주소를 담는 리스트
call=[]       # 전화번호를 담는 리스트

for i in range(1, 54):
    html=urlopen('https://www.hollys.co.kr/store/korea/korStore2.do?pageNo='+str(i)+'&sido=&gugun=&store=')
    soup=BeautifulSoup(html, 'html.parser')

    tbodys = soup.find('tbody')
    tbody = tbodys.find_all('td')

    for t in tbody:
        tds.append(t.text)

for i in range(1, len(tds) + 1):
    if (i % 6 in [0, 1, 2, 4]):
        result.append(tds[i - 1])



for i in range(len(result)):
        if i%4==0:
            region.append(result[i])
        elif i%4==1:
            shop_name.append(result[i])
        elif i%4==2:
            shop_addr.append(result[i])
        else:
            call.append(result[i])


region_df=pd.DataFrame(region, columns=['지역'])
shop_name_df=pd.DataFrame(shop_name, columns=['매장이름'])
shop_addr_df=pd.DataFrame(shop_addr, columns=['매장주소'])
call_df=pd.DataFrame(call, columns=['전화번호'])
info=pd.concat([region_df, shop_name_df, shop_addr_df, call_df], axis=1)

pd.DataFrame(info)
```
