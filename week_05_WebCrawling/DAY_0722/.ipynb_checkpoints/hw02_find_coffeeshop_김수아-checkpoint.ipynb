# 2. 저장된 csv 파일을 읽어서 사용자가 입력한 지역에 있는 매장 정보 출력

```python
import pandas as pd

shop_info=pd.read_csv('hollys_branches.csv')
shop_info
```

```python
# 각 데이터프레임을 리스트로 만들기
region_list=shop_info['지역'].tolist()
name_list=shop_info['매장명'].tolist()
addr_list=shop_info['주소'].tolist()
call_list=shop_info['전화번호'].tolist()
```

```python
city=input('검색할 매장의 도시를 입력하세요 : ')
number=1  # 순번을 표시할 숫자

for i in range(len(region_list)):
    if city in region_list[i]:
        print("[ {}]: ['{}', '{}']".format(number, addr_list[i], call_list[i]))
        number+=1  # 순번이 하나씩 늘어나도록
```
