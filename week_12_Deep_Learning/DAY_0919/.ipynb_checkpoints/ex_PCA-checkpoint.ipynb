## PCA(주성분 분석)
---
- 전체 특성(Feature)에서 주요 특성(Feature) 압축/융합하는 비지도 학습
- fit() + transform() 2단계 거쳐 주성분 분석

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import scatter_matrix
```

## [1] 데이터 준비

```python
X, y=load_iris(return_X_y=True, as_frame=True)
```

```python
type(X), type(y)
```

```python
scatter_matrix(X,figsize=(12, 10))
plt.tight_layout()
plt.show()
```

```python
X
```

```python
irisDF=pd.concat([X, y], axis=1)
```

```python
target_0=irisDF[irisDF['target']==0]
target_1=irisDF[irisDF['target']==1]
target_2=irisDF[irisDF['target']==2]
```

```python
plt.scatter(target_0['sepal length (cm)'], target_0['sepal width (cm)'], c='orange')
plt.scatter(target_1['sepal length (cm)'], target_1['sepal width (cm)'], c='blue')
plt.scatter(target_2['sepal length (cm)'], target_2['sepal width (cm)'], c='red')
```

## 데이터 전처리

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(X)
```

```python
X_scaled=scaler.transform(X)
```

## PCA(주성분 분석)
---
- iris의 4개 Feature 중에서 주성분 데이터 추출

```python
pca=PCA()
```

```python
# 지정된 데이터에서 주성분 추출
pca.fit(X_scaled)
```

```python
# 주성분 행렬
pca.singular_values_
```

```python
# 주성분에 투영 후 분산 정도 설명
print(pca.explained_variance_)
print(pca.explained_variance_ratio_)
```

## 주성분 개수 지정
---
- 방법1)
    * explained_variance_ratio_의 누적값에서 원하는 분포 정도 지정해서 PC 개수 계산
- 방법2)
    * n_components = 0.0 ~ 1.0 : 주성분에 투영된 데이터 분포 정도 지정하면 PC 개수 자동 계산

### 방법[1] : explained_variance_ratio_ 누적값으로 PC 차원 수 계산

```python
cumsum=np.cumsum(pca.explained_variance_ratio_)
d=np.argmax(cumsum>=0.95)+1

print(f'주성분 개수(차원) : {d}')
```

```python
## 지정된 PC 개수로 주성분 데이터 압축/융합
pca=PCA(n_components=d)
```

```python
pca.fit(X_scaled)
```

```python
irisPCA=pca.transform(X_scaled)
```

```python
irisPCA.shape
```

```python
irisPCA[:5]
```

```python
## 주성분과 데이터 투영 분산 정도 시각화
plt.plot(cumsum)
plt.plot(d, 0.95, 'ro')
plt.plot([2.0, 2.0], [0, 0.95], 'b--')
plt.axis([0, 4, 0, 1])
plt.xlabel('Dimension')
plt.ylabel('Explained Variance')
```

### 방법[2] : n_components = 비율값

```python
pca3=PCA(n_components=0.95)
```

```python
pca3.fit(X_scaled)
```

```python
print(pca3.explained_variance_)
print(pca3.explained_variance_ratio_)
```

```python
irisPCA3=pca3.transform(X_scaled)
irisPCA3.shape
```

```python
irisPCA3_DF=pd.DataFrame(irisPCA3, columns=['pca_1', 'pca_2'])
```

```python
irisPCA3_DF
```

```python
irisPCA_DF=pd.concat([irisPCA3_DF, y], axis=1)
```

```python
irisPCA_DF
```

```python
# 타겟별 시각화
target_0=irisPCA_DF[irisPCA_DF['target']==0]
target_1=irisPCA_DF[irisPCA_DF['target']==1]
target_2=irisPCA_DF[irisPCA_DF['target']==2]
```

```python
plt.scatter(target_0['pca_1'], target_0['pca_2'], c='orange')
plt.scatter(target_1['pca_1'], target_1['pca_2'], c='blue')
plt.scatter(target_2['pca_1'], target_2['pca_2'], c='red')
plt.legend(['target_0', 'target_1', 'target_2'])
plt.show()
```

```python
# PC 개수 추출하기 전으로 되돌리기
re=pca3.inverse_transform(irisPCA)
```
