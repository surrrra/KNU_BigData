###  모듈 로딩

```python
from scikeras.wrappers import KerasRegressor, KerasClassifier
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, InputLayer
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.utils import set_random_seed
from tensorflow.keras.backend import clear_session
import numpy as np
```

```python
## 동일 결과 출력 시드 설정
```

```python
clear_session()
np.random.seed(42)
set_random_seed(42)
```

### [1] 데이터 로딩 및 준비

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
x_data=data.data
y=data.target
```

```python
x_data.shape, y.shape
```

### [2] 모델 생성

```python
# 파라미터 => KerasClassifier 또는 KerasRegressor의 parameter로 추가됨
#def build_model(neuron_1, neuron_2, neuron_3, learning_rate, input_shape=(4,)):
def build_model(n_hidden, neuron_1, neuron_2, neuron_3, learning_rate, input_shape=(4,)):    
    model = Sequential()
    
    model.add(InputLayer(input_shape=input_shape))
    
#     for layer in range(n_hidden):
#         model.add(Dense(n_neurons, activation="relu"))
    model.add(Dense(neuron_1, activation="relu"))
    model.add(Dense(neuron_2, activation="relu"))
    model.add(Dense(neuron_3, activation="relu"))
        
    model.add(Dense(3, activation='softmax'))
    
    return model
```

for layer in range(n_hidden):
    model.add(Dense(<b>n_neurons</b>, activation='relu'))

n_neurons : 이 for문이 은닉층 개수를 몇 개 해야할지 모르겠을 때 넣는 거였는데 n_neurons는 원래 함수 들어올 때 파라미터로 받는 거였음  
def build_model(n_hidden, (neuron_1, 2, 3 대신 n_neurons), ...)  

```python
keras_clf = KerasClassifier(build_model, 
                            n_hidden=1, 
                            neuron_1=30, 
                            neuron_2=50, 
                            neuron_3=20, 
                            learning_rate=0.001,
                            loss="sparse_categorical_crossentropy",
                            epochs=5)
```

```python
keras_clf.fit(x_data, y)
```

```python
mse_test = keras_clf.score(x_data, y)
```

```python
print(f'mse_test ; {mse_test}')
```

```python
keras_clf.get_params
```

### [3]Hyperparameter Tunning

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    "n_hidden": [1,3,5,10],
    "neuron_1": [10,30,60],
    "neuron_2": [30, 60, 90],
    "neuron_3": [10,50,70],
    "optimizer": ["adam", "sgd"],
    "learning_rate": [0.0001, 0.001, 0.1],
}
gs = GridSearchCV(keras_clf, params, 
                  refit=False, cv=3, 
                  scoring='accuracy',
                  verbose=False)
```

```python
print(f'fitting....')
gs.fit(x_data, y)
print('EMD')
```

```python
print(gs.best_score_, gs.best_params_, sep='\n')
```
