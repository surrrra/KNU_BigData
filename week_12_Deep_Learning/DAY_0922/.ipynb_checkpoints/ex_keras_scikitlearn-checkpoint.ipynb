```python
from scikeras.wrappers import KerasClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, InputLayer
import numpy as np
import pandas as pd
```

```python
X, y=load_iris(as_frame=True, return_X_y=True)
```

```python
X
```

```python
y
```

```python
def get_model(hidden_layer_dim, meta):
    
    n_features_in_=meta['n_features_in_']
    X_shape_=meta['X_shape_']
    n_classes_=meta['n_classes_']
    
    model=Sequential()
    model.add(Dense(n_features_in_, input_shape=X_shape_[1:], activation='relu'))
    model.add(Dense(hidden_layer_dim, activation='relu'))
    model.add(Dense(n_classes_, activation='softmax'))
    
    return model
    
```

```python
clf=KerasClassifier(
get_model,
loss='categorical_crossentropy',
hidden_layer_dim=100,)
```

```python
clf.fit(X, y)
```

```python
y_proba=clf.predict_proba(X)
```

```python
params={
    'hidden_layer_dim':[32, 50, 64, 100, 128, 150, 200],
    'loss':['categorical_crossentropy'],
    'optimizer':['adam', 'sgd'],
    'optimizer__learning_rate':[0.0001, 0.001, 0.01, 0.1],
}

gs=GridSearchCV(clf, params, refit=False, cv=3, scoring='accuracy')
```

```python
gs.fit(X, y)
```

```python
print(gs.best_score_)
print(gs.best_params_)
```

```python
def build_model(n_hidden, n_neurons, learning_rate, input_shape=(4, )):
    model=Sequential()
    model.add(InputLayer(input_shape=input_shape))
    
    for layer in range(n_hidden):
        model.add(Dense(n_neurons, activation='relu'))
        
    model.add(Dense(3, activation='softmax'))
    
    return model
```
