## K-Means 분할적 군집화
---
- 지정된 k만큼 임의의 중심점(Centroid)를 지정
- 중심점(Centroid)를 기준으로 가까운 거리에 있는 데이터를 하나의 군집으로 묶기
- 군집마다 다시 중심점(Centroid) 계산
- 중심점(Centroid) 이동이 더이상 변경되지 않을 때까지 반복하는 방법

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

### [1] 데이터 로딩 및 준비

```python
iris=load_iris()
```

```python
irisData=iris.data
```

```python
irisData.shape
```

```python
irisData=irisData[:, :2]
irisData.shape
```

### [2] 데이터 처리

```python
irisCluster=KMeans(n_clusters=2)
```

```python
irisCluster.fit(irisData)
```

```python
# Cluster 객체 속성 확인
print(f'cluster_centers_ : {irisCluster.cluster_centers_}')
print(f'labels_ : {irisCluster.labels_}')
print(f'labels_ Counts : {np.unique(irisCluster.labels_, return_counts=True)}')
print(f'inertia_ : {irisCluster.inertia_}')
print(f'n_iter_ : {irisCluster.n_iter_}')
```

```python
def check_kmeans(num):

    irisCluster=KMeans(n_clusters=num)
    irisCluster.fit(irisData)
    
    # 샘플 데이터 => 결정된 군집으로 변환
    irisCluster.transform(irisData)
    
    print('='*70)
    print(f'num = {num}')
    # Cluster 객체 속성 확인
    print(f'cluster_centers_ : {irisCluster.cluster_centers_}')
    print(f'labels_ : {irisCluster.labels_}')
    print(f'labels_ Counts : {np.unique(irisCluster.labels_, return_counts=True)}')
    print(f'inertia_ : {irisCluster.inertia_}')
    print(f'n_iter_ : {irisCluster.n_iter_}')
    print()
```

```python
for i in range(1, 6):
    print(i)
    check_kmeans(i)
```

```python
# 샘플 데이터 => 결정된 군집으로 변환
irisCluster.transform(irisData)
```

## 시각화
---

```python
plt.scatter(irisData[:, 0], irisData[:, 1], c='skyblue')
plt.plot(irisCluster.cluster_centers_[0][0], irisCluster.cluster_centers_[0][1], 'ro', label='c1')
plt.plot(irisCluster.cluster_centers_[1][0], irisCluster.cluster_centers_[1][1], 'go', label='c2')
plt.show()
```

---

```python
from sklearn.metrics import silhouette_score

silhouette_score(irisData, irisCluster.labels_, metric='euclidean')
```

```python
plt.scatter(irisData[:, 0], irisData[:, 1], c=irisCluster.labels_)
for idx in range(irisCluster.n_clusters):
    plt.plot(irisCluster.cluster_centers_[idx][0], irisCluster.cluster_centers_[idx][1], 'ro')
plt.show()
```

```python
inerList=[]
for k in range(1, 11):
    km=KMeans(n_clusters=k, random_state=11)
    km.fit(irisData)
    inerList.append(km.inertia_)
    
    plt.scatter(irisData[:, 0], irisData[:, 1], c=km.labels_)
    for idx in range(km.n_clusters):
        plt.plot(km.cluster_centers_[idx][0], km.cluster_centers_[idx][1], 'ro')
    plt.show()
```
