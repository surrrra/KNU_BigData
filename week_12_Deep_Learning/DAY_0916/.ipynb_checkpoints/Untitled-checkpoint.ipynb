```python
from tensorflow.keras.utils import set_random_seed, plot_model
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv2D, Flatten
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import PIL.Image as pilimg
```

```python
def open_noisy_image(num):
    image=pilimg.open('./file/noisy/'+str(num)+'.png')
    pix=np.array(image)
    
    return pix
```

```python
open_noisy_image(0).reshape(-1, 420*540)
```

```python
def open_clean_image(num):
    image=pilimg.open('./file/clean/'+str(num)+'.png')
    pix=np.array(image)
    
    return pix
```

```python
open_clean_image(0).shape
```

```python
noisy_pix=open_noisy_image(0)/255.0
noisy_pix=noisy_pix.reshape(-1, 420*540)
```

```python
clean_pix=open_clean_image(0)/255.0
clean_pix=clean_pix.reshape(-1, 420*540)
```

```python
noisy_train=np.array([])
noisy_test=np.array([])
clean_train=np.array([])
clean_test=np.array([])

for i in range(172):
    noisy_pix=open_noisy_image(i)
    noisy_train.append(noisy_pix/255.0)
    
    clean_pix=open_clean_image(i)
    clean_train.append(clean_pix/255.0)
    
for i in range(173, 216):
    noisy_pix=open_noisy_image(i)
    noisy_test.append(noisy_pix/255.0)
    
    clean_pix=open_clean_image(i)
    clean_test.append(clean_pix/255.0)
```

```python
len(noisy_train), len(noisy_test), len(clean_train), len(clean_test)
```

```python
def create_model(num):
    model=Sequential(name='model_'+str(num))
    model.add(Dense(units=num, activation='relu', input_shape=(420*540, )))
    model.add(Dense(units=num, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(420*540, activation='sigmoid'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    return model
```

```python
model=create_model(100)

his=model.fit(noisy_train, clean_train, epochs=10)
```
