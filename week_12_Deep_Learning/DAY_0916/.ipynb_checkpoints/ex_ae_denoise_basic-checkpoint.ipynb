## AutoEncode 노이즈 제거
---
- 데이터에 노이즈 제거로 활용
- 데이터에서 주요 특징만 추출 압축하기 때문에 노이즈는 주요 특징에서 제거
- 압축된 주요 특징으로 데이터 복원 => 노이즈 제거

```python
from tensorflow.keras.datasets.mnist import load_data
from tensorflow.keras.utils import set_random_seed, plot_model
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv2D, Flatten
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import PIL.Image as pilimg
```

```python
(x_train, y_train), (x_test, y_test)=load_data()
```

```python
x_train.shape, x_test.shape
```

```python
x_train=x_train.reshape(-1, 28*28)
x_test=x_test.reshape(-1, 28*28)
```

```python
x_train=x_train/255.0
x_test=x_test/255.0
```

```python
# 데이터셋에 노이즈 추가
# np.random.normal(평균loc=0.0, 표준편차scale=1.0, 출력사이즈size=None)
x_train_noised=x_train+np.random.normal(0, 0.5, size=x_train.shape)
x_test_noised=x_test+np.random.normal(0, 0.5,size=x_test.shape)

# 픽셀 범위 0.0~1.0
# np.clip(array, min, max)
x_train_noised=np.clip(x_train_noised, a_min=0, a_max=1)
x_test_noised=np.clip(x_test_noised, a_min=0, a_max=1)
```

```python
plt.imshow(x_train[0].reshape(28, 28), cmap='gray')
plt.show()
```

```python
# 노이즈 처리 된 사진 확인
plt.imshow(x_train_noised[0].reshape(28, 28), cmap='gray')
plt.show()
```

```python
# W, b 초기값 고정
set_random_seed(11)
```

```python
def myModel(num):
    '''
    model 생성하는 함수
    '''
    
    model=Sequential(name='model_'+str(num))
    model.add(Dense(units=num, activation='relu', input_shape=(784, )))
    model.add(Dense(784, activation='sigmoid'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    
    return model
```

```python
def study(model, num):
    '''
    학습하고 예측,
    노이즈 처리된 사진과 예측(노이즈 제거)한 사진 출력 함수
    '''
    
    model.fit(x_train_noised, x_train, epochs=10, verbose=0)
    print(model.evaluate(x_test, x_test))
    pre_out=model.predict(x_train_noised[num].reshape(-1, 784))
    
    plt.imshow(x_train_noised[num].reshape(28, 28), cmap='gray')
    plt.show()
    
    plt.imshow(pre_out.reshape(28, 28), cmap='gray')
    plt.show()
```

```python
# unit 10개로 x_train_noised[0]의 노이즈 제거(예측)
study(myModel(10), 0)
```

```python
study(myModel(100), 0)
```

```python
study(myModel(100), 5)
```

```python
study(myModel(100), 9)
```

```python
study(myModel(100), 8)
```

```python
study(myModel(100), 7)
```

### 그림판에 직접 쓴 이미지 불러와서 예측

```python
model=myModel(150)
model.fit(x_train_noised, x_train, epochs=10, verbose=0)

image = pilimg.open('my_four.jpg')
image=image.convert('L')
pix = np.array(image)/255.0

pix_noised=pix+np.random.normal(0, 0.5, size=pix.shape)
pix_noised=np.clip(pix_noised, a_min=0, a_max=1)

pre_out=model.predict(pix_noised.reshape(-1, 784))

plt.imshow(pix.reshape(28, 28), cmap='gray')   # 원본 사진
plt.show()

plt.imshow(pix_noised.reshape(28, 28), cmap='gray')  # 노이즈 처리된 사진
plt.show()

plt.imshow(pre_out.reshape(28, 28), cmap='gray')  # 노이즈 처리된걸 예측(노이즈 제거)한 사진
plt.show()
```

```python
# 예측

image = pilimg.open('my_five.jpg')
image=image.convert('L')
pix = np.array(image)/255.0

pix_noised=pix+np.random.normal(0, 0.5, size=pix.shape)
pix_noised=np.clip(pix_noised, a_min=0, a_max=1)

model=myModel(150)
model.fit(x_train_noised, x_train, epochs=10, verbose=0)

pre_out=model.predict(pix_noised.reshape(-1, 784))

plt.imshow(pix.reshape(28, 28), cmap='gray')
plt.show()

plt.imshow(pix_noised.reshape(28, 28), cmap='gray')
plt.show()

plt.imshow(pre_out.reshape(28, 28), cmap='gray')
plt.show()
```
