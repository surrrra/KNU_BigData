```python
import csv
import matplotlib.pyplot as plt
import platform

city_list = ['','','','','', '', '']    # csv   

################################################################################################

f = open('gender.csv', encoding = 'euc_kr')
data = csv.reader(f)
header = next(data)
# row[0]: 

male_list = []      
female_list = []

for city in city_list:
    for row in data: 
        district = ' '
        if district in row[0]: # ''  
            male_list.append(int(row[104].replace(',', '')))
            
            female_list.append(int(row[207].replace(',', '')))
        break
                
    for i in range(7):
              
        for row in data: 
            district = ' ' + city_list[i]
            if district in row[0]: # ' + OO'   
                male_list.append(int(row[104].replace(',', '')))
                female_list.append(int(row[207].replace(',', '')))
#                 print(male_list, type(male_list))
#                 print(female_list, type(female_list))
                break   
                #          
                # male_list female_list    ( 8 )
            
#         for row in data: 
#                 district = ' '
#                 if district in row[0]: # ''  
#                     male_list.append(int(row[104].replace(',', '')))
#                     female_list.append(int(row[207].replace(',', '')))
#                 break
#     print(len(male_list),len(female_list))                
f.close()

################################################################################################

# for        
def draw_piechart(male_list, female_list):
    
    if platform.system() == 'Windows':
        plt.rc('font', family='Malgun Gothic')
    else:
        plt.rc('font', family='AppleGothic')
                                                               # (, , )   
    fig, axes = plt.subplots(2,4, figsize=(20,14))             #   . for     
    fig.suptitle('    ', fontsize=25)  #  fig  
    
    for i in range(8):
        population = [male_list[i], female_list[i]]
        color = ['tomato', 'royalblue']
        
        plt.subplot(2, 4, i+1)      # plt.subplot(, , )
        plt.pie(population, labels = ['',''], autopct='%.1f%%',
                colors=color, startangle=90)
        plt.title(' ' + city_list[i], fontsize=20)
        plt.rc('font', size=15)     #    

    plt.show()   # plt.show() 1    

################################################################################################

city_list = ['','','','','', '', '', '']

draw_piechart(male_list, female_list)
#  ,      
#   -  x8   ,      
```
