```python
import csv
import matplotlib.pyplot as plt
import platform

city_list = ['중구','동구','서구','남구','북구', '수성구', '달서구']    # csv 특성상 순서대로 적어줘야

################################################################################################

f = open('gender.csv', encoding = 'euc_kr')
data = csv.reader(f)
header = next(data)
# row[0]: 행정구역

male_list = []      
female_list = []

for city in city_list:
    for row in data: 
        district = '대구광역시 '
        if district in row[0]: # '대구광역시' 자료만 출력
            male_list.append(int(row[104].replace(',', '')))
            
            female_list.append(int(row[207].replace(',', '')))
        break
                
    for i in range(7):
              
        for row in data: 
            district = '대구광역시 ' + city_list[i]
            if district in row[0]: # '대구광역시 + OO'이 포함된 자료만 출력
                male_list.append(int(row[104].replace(',', '')))
                female_list.append(int(row[207].replace(',', '')))
#                 print(male_list, type(male_list))
#                 print(female_list, type(female_list))
                break   
                # 특정 도시의 데이터 중에서 제일 먼저 나오는 데이터만 분석하기 위함
                # male_list와 female_list는 필요한 자료로 차있는 상태( 8개씩 )
            
#         for row in data: 
#                 district = '대구광역시 '
#                 if district in row[0]: # '대구광역시' 자료만 출력
#                     male_list.append(int(row[104].replace(',', '')))
#                     female_list.append(int(row[207].replace(',', '')))
#                 break
#     print(len(male_list),len(female_list))                
f.close()

################################################################################################

# for문 돌기 전 그래프 그리는 것만 함수로 따로 빼놓기
def draw_piechart(male_list, female_list):
    
    if platform.system() == 'Windows':
        plt.rc('font', family='Malgun Gothic')
    else:
        plt.rc('font', family='AppleGothic')
                                                               # (행, 열, 크기) 순으로 입력하는 것
    fig, axes = plt.subplots(2,4, figsize=(20,14))             # 그래프 하나하나를 의미. for문 밖으로 빼서 한번만 적어줘야 함
    fig.suptitle('대구광역시 구별 남녀 인구 비율', fontsize=25)  # 받은 fig에 대한 것임
    
    for i in range(8):
        population = [male_list[i], female_list[i]]
        color = ['tomato', 'royalblue']
        
        plt.subplot(2, 4, i+1)      # plt.subplot(행, 열, 위치)
        plt.pie(population, labels = ['남성','여성'], autopct='%.1f%%',
                colors=color, startangle=90)
        plt.title('대구광역시 ' + city_list[i], fontsize=20)
        plt.rc('font', size=15)     # 파이차트 안에서의 글자 크기

    plt.show()   # plt.show()는 1번만 호출하면 되니까 위치 중요

################################################################################################

city_list = ['중구','동구','서구','남구','북구', '수성구', '달서구', '전체']

draw_piechart(male_list, female_list)
# 함수 호출, 받는 인수들 표시 위해 파라미터 적어주고
# 구별로 비교중 - 각각 x8 하는 셈 , 데이터 하나에 대해 구가 맞는지만 
```
