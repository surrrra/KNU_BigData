```python
# 모듈 로딩 -------------------
import csv                        # csv 파일 읽기 위해
import matplotlib.pyplot as plt   # 그래프(파이차트) 그리기 위해
import platform                   # 컴퓨터 OS(운영체제)를 확인하는 위해

# csv파일 읽기
f=open('gender.csv', encoding='euc-kr')
data=csv.reader(f)

# 변수 설정
city_list = ['대구광역시', '대구광역시 중구', '대구광역시 동구', '대구광역시 서구', '대구광역시 남구', '대구광역시 북구', '대구광역시 수성구', '대구광역시 달서구']
i=8   # 그래프 위치 조절을 위해

# 운영체제에 따른 그래프 font 설정
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='AppleGothic')    

# 여러 그래프 한 번에 그리기
fig, axes = plt.subplots(2,4,figsize=(10,7))
fig.suptitle('대구광역시 구별 남녀 인구 비율', fontsize=20)   

for city in city_list:
    male_count=0     # 초기화
    female_count=0   # 초기화    
    for row in data: 
        if city in row[0]:
            male_count=int(row[104].replace(',',''))  # 세자리수 ,제거 및 숫자 변환
            female_count=int(row[104+103].replace(',',''))
            break

#     print(f"{city} 남자 인구수: {male_count:,}명, 여자 인구수: {female_count:,}명")
    population = [male_count, female_count]    
    plt.subplot(2,4,i)    
    plt.pie(population, labels=['남성','여성'], autopct='%.1f%%', startangle=90)
    plt.title(city)
#     print(i)
    i=i%8+1

plt.show()            
f.close()
```
