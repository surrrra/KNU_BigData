# 과제3
### 대구시의 7개 구별(중구, 동구, 서구, 남구, 북구, 수성구, 달서구) 남녀 비율 및 
### 대구시 전체의 남녀 비율을 각각의 파이 차트로 나타내시오.
- subplots를 이용하여 4x2 형태의 총 8개의 subplot을 파이 차트로 구현
- gender.csv 사용

```python
import csv
import matplotlib.pyplot as plt
import platform

# 대구광역시 각 구의 인구 수를 구하는 함수
def get_population(gu):
    f=open('gender1.csv', encoding='euc_kr')
    data=csv.reader(f)
    population=[]  # 남, 여 인구 값을 담는 리스트
    male_count=0
    female_count=0
    
    for row in data:
        if '대구광역시 ' +gu in row[0]:  # '대구광역시 중구' 이런 식의 값을 찾음
            male_count=int(row[104].replace(',',''))  # 남자 총 인원 수
            female_count=int(row[207].replace(',',''))  # 여자 총 인원 수
            break   # 제일 먼저 등장하는 값'만' 찾으려고
        
    #print('{} 남자 인구 수 : {}명, 여자 인구 수 : {}명'.format(gu,format(male_count,','),format(female_count,',')))     

    population=[male_count, female_count]
    return population

pop=get_population('중구')
print(pop)
```

```python
# 파이차트 여러 개 그리는 함수
def draw_sub_pie():
    fig, axes=plt.subplots(2, 4, figsize=(15, 6), sharex=True, sharey=True)
    fig.suptitle('대구광역시 구별 남녀 인구 비율', fontsize=15)
    
    if platform.system()=='Windows':
        plt.rc('font', family='Malgun Gothic')
    else:
        plt.rc('font', family='AppleGothic')

    color=['cornflowerblue', 'tomato']

    axes[0, 0].pie(get_population('중구'), labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)
    axes[0, 0].set_title('대구광역시 중구')
    
    axes[0, 1].pie(get_population('동구'), labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)
    axes[0, 1].set_title('대구광역시 동구')
    
    axes[0, 2].pie(get_population('서구'), labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)
    axes[0, 2].set_title('대구광역시 서구')
    
    axes[0, 3].pie(get_population('남구'), labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)
    axes[0, 3].set_title('대구광역시 남구')
    
    axes[1, 0].pie(get_population('북구'), labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)
    axes[1, 0].set_title('대구광역시 북구')
    
    axes[1, 1].pie(get_population('수성구'), labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)
    axes[1, 1].set_title('대구광역시 수성구')
    
    axes[1, 2].pie(get_population('달서구'), labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)
    axes[1, 2].set_title('대구광역시 달서구')
    
    axes[1, 3].pie(get_population(''), labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)
    axes[1, 3].set_title('대구광역시 전체')
    
draw_sub_pie()
```
