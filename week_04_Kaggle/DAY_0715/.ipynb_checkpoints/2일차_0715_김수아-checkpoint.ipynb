# 공공데이터 과제 2
### 지하철 각 노선별 최대 하차 인원을 막대그래프로 표시하고, 하차인원 출력
- 출근 시간대 : 07:00~08:59
- 각 지하철 노선별 가장 많이 내리는 지하철 역 분석
- 하차 인원은 1,000 단위로 콤마를 찍어서 구분
- 6개의 지하철 역을 막대 그래프로 표시
- Bar chart의 x축은 (노선+지하철 역이름), y축은 인원수
- platform 모듈을 사용하여 한글폰트 지정

```python
import pandas as pd
import matplotlib.pyplot as plt
import platform

df=pd.read_csv('subwaytime.csv', encoding='CP949', header=[0, 1])
df.head()
```

```python
# 호선명, 지하철역, 7시대 하차, 8시대 하차 컬럼 추출
commute_df=df.iloc[:, [1, 3, 11, 13]]
commute_df.head()
```

```python
# 멀티컬럼 이름 바꾸기
commute_df.columns=[['호선명', '지하철역', '07:00:00~07:59:59', '08:00:00~08:59:59'], ['Unnamed: 1_level_1', 'Unnamed: 3_level_1','하차', '하차']]
commute_df.head()
```

```python
# 컬럼 이름 잘 바뀌었는지 확인
commute_df
```

```python
# 컬럼 타입 확인
commute_df.dtypes
```

```python
result=[]  # 지하철 노선별 최대 하차 인원 담는 리스트
x_ticks=[]  # x축 라벨을 위한 리스트

for i in range(1, 8):
    commute_line_df=commute_df[commute_df[('호선명', 'Unnamed: 1_level_1')]==str(i)+'호선']   # 호선 체크
    sum_line_df=commute_line_df.sum(axis=1)  # 각 호선의 역마다 7시~8시 59분까지 하차 인원 합
    
    max_line_num=sum_line_df.max()  # 각 호선의 최대 하차 인원
    result.append(max_line_num)
    max_line_idx=sum_line_df.idxmax()  # 각 호선의 최대 하차 인원의 인덱스
    max_line, max_station=df.iloc[max_line_idx, [1, 3]]  # 원래의 데이터프레임에서 인덱스가 max_line_idx인 행의 호선, 역 이름 저장
    x_ticks.append('{}:{}'.format(max_line, max_station))   # 1호선:종각 의 형태
    print('출근 시간대 {} 최대 하차역은 {}역이고 하차 인원은 {:,}명입니다.'.format(max_line, max_station, max_line_num))
    
    
    
plt.figure(figsize=(8, 6))
plt.bar(range(7), result)
plt.xticks(range(7), x_ticks, rotation=80)
plt.title('출근 시간대 지하철 노선별 최대 하차 인원 및 하차역')


# 한글 폰트 지정
if platform.system()=='Windows':
        plt.rc('font', family='Malgun Gothic', size=10)
else:
    plt.rc('font', family='AppleGothic', size=8)
```
